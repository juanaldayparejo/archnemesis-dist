In archNEMESIS each gaseous species can be defined as a mixture of isotopes with relative abundances representative of the Earth's atmosphere (i.e., ISO=0), or as individual isotopologues (i.e., ISO>0). In this page we provide a complete list of all isotopologues included in the archNEMESIS database.
